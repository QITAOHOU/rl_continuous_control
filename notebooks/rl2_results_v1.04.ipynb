```python
import gym
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = "/Volumes/BC_Clutch/Dropbox/DeepRLND/rl_continuous_control/"
```

```python
def calc_running_avg(data):
    running_avg = []
    for i,ep in enumerate(data):
        progress = data[:i+1]
        length = min(100, len(progress))
        average = np.mean(progress[-length:])
        running_avg.append(average)
    return running_avg

def plot_results(data,env_name,agent_name, timestamp):
    episode_count = len(data)
    print(f"Trained in {episode_count} Episodes")
    plt.plot(np.arange(len(data)), data, label="By Episode")
    plt.plot(np.arange(len(data)), calc_running_avg(data), label="Running Average")
    plt.ylabel('Reward')
    plt.xlabel('Episode')
    plt.title(f"{env_name}, {agent_name}: Average Agent Reward")
    plt.legend()
    plt.savefig(PATH + f"charts/{timestamp}_plotresults.png")
```

### Reacher20, DDPG

```python
data = (pd.read_pickle(PATH + "results/201901030755_Reacher20_ResultDict.pkl")
        [('Reacher20','DDPG')]['Scores']['Reacher20','DDPG']['Scores'])
plot_results(data, "Reacher20", "DDPG","201901030755")
```
